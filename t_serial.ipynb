```python
import serial
import eventlet
from eventlet import tpool, backdoor
import random

class SerialSocket(serial.Serial):
    def recv(self):
        # 等待
        data = tpool.execute(self.read, 1)
        # 读完
        while self.in_waiting:
            data += tpool.execute(self.read, self.in_waiting)
        return data
```

```python
com2 = SerialSocket(port="COM1", baudrate=115200)
com1 = SerialSocket(port="COM2", baudrate=115200)

c = random.randbytes(10000)
print (c)
```
